```python
import json
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb
import numpy as np
import pandas as pd
from pprint import pprint
from datetime import datetime
```

```python
with open('results-latest.json', 'r') as r:
    results = json.load(r)['results']
with open('profiles-latest.json', 'r') as p:
    profiles = json.load(p)['profiles']
print "{} results loaded".format(len(results))
```

```python
minAppVersionCode = 50
parametersVersion = '3.1-production-2'
min_n_results = 10
good_profiles = [p for p in profiles
                 if p['profile_data'].get('appVersionCode') >= minAppVersionCode
                 and '+' not in p['profile_data'].get('appVersionName')
                 and p['n_results'] >= min_n_results
                 and p['profile_data'].get('parametersVersion') >= parametersVersion]
good_profile_ids = [p['id'] for p in good_profiles]
print ("{n_profiles} profiles worth keeping (>= {min_n_results} results, "
       "appVersionCode >= {minAppVersionCode}, parametersVersion >= {parametersVersion})").format(
            n_profiles=len(good_profiles), min_n_results=min_n_results, 
            minAppVersionCode=minAppVersionCode, parametersVersion=parametersVersion)
```

```python
good_results = [r for r in results if r['profile_id'] in good_profile_ids]
print "{} results uploaded by those good profiles".format(len(good_results))
```

```python
ordered_profiles = sorted(good_profiles, key=lambda pr: pr['n_results'], reverse=True)
```

```python
isodate = r'%Y-%m-%dT%H:%M:%S.%fZ'
profile_durations = []
profile_probe_frequencies = []
profile_mornings = []
profile_evenings = []

for p in ordered_profiles:
    n_processed = 0
    types = {}
    dates = []
    for r in good_results:
        if r['profile_id'] != p['id']:
            continue
        
        dates.append(r['created_at'])
        rdata = r['result_data']
        
        # Got a location
        if 'locationLatitude' in rdata:
            if 'location' not in types:
                types['location'] = 0
            types['location'] += 1
        
        # Got a probe/questionnaire
        elif 'pageGroups' in rdata:
            tipe = rdata['type']
            status = rdata['status']
            if tipe not in types:
                types[tipe] = {}
            if status not in types[tipe]:
                types[tipe][status] = 0
            types[tipe][status] += 1
        n_processed += 1

    
    title = "## Profile {} ({}/{} results processed)".format(p['id'][:10] + '...', n_processed, p['n_results'])
    print '-' * len(title)
    print title
    print
    
    min_date = min(dates)
    max_date = max(dates)
    span_hours = (datetime.strptime(max_date, isodate)
                 - datetime.strptime(min_date, isodate)).total_seconds() / 3600.0
    span_days = span_hours / 24.0
    profile_durations.append(span_days)
    profile_probe_frequencies.append(sum(types.get('probe', {}).values()) / span_days)
    profile_mornings.append(sum(types.get('morningQuestionnaire', {}).values()))
    profile_evenings.append(sum(types.get('eveningQuestionnaire', {}).values()))
    
    print " # From {} to {} ({} days)".format(min(dates)[:10], max(dates)[:10], round(span_days, 1))
    print
    for tipe, statuses in sorted(types.iteritems(), key=lambda item: item[0]):
        if isinstance(statuses, int):
            print "   {}: {} (~{} per day / ~{} per hour)".format(tipe, statuses, 
                                                                  round(statuses / span_days, 1),
                                                                  round(statuses / span_hours, 1))
        else:
            total = sum(statuses.values())
            print "   {}: {} (~{} per day)".format(tipe, total, round(total / span_days, 1))
            for status, n in statuses.iteritems():
                print "     {}: {}".format(status, n)
        print
    print
    print
profile_durations = np.array(profile_durations)
profile_probe_frequencies = np.array(profile_probe_frequencies)
profile_mornings = np.array(profile_mornings)
profile_evenings = np.array(profile_evenings)
```

```python
fig, axs = plt.subplots(2, 2, sharex=True, figsize=(12, 8))
palette = sb.color_palette()

axs[0, 0].set_axis_off()

axs[1, 0].scatter(profile_durations, profile_probe_frequencies, color=palette[0])
axs[1, 0].set_title("#probes / day vs. #days the profile has participated")
axs[1, 0].set_xlabel("#days participated")
axs[1, 0].set_ylabel("#probes / day")

axs[0, 1].scatter(profile_durations, profile_mornings, color=palette[1])
axs[0, 1].plot([profile_durations.min(), profile_durations.max()], 
               [profile_durations.min(), profile_durations.max()],
               '--', color=palette[1] + (0.3,))
axs[0, 1].set_title("#morningQuestionnaires vs. #days the profile has participated")
axs[0, 1].set_ylabel("#morningQuestionnaires")

axs[1, 1].scatter(profile_durations, profile_evenings, color=palette[2])
axs[1, 1].plot([profile_durations.min(), profile_durations.max()], 
               [profile_durations.min(), profile_durations.max()],
               '--', color=palette[2] + (0.3,))
axs[1, 1].set_title("#eveningQuestionnaires vs. #days the profile has participated")
axs[1, 1].set_xlabel("#days participated")
axs[1, 1].set_ylabel("#eveningQuestionnaires")

plt.tight_layout()
```
